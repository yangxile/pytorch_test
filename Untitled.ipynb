```python
import numpy as np
import matplotlib.pyplot as plt

# 创建一个随机矩阵
matrix = np.random.rand(6, 6)

# 定义一个颜色映射，这里使用'viridis'，你也可以选择其他颜色映射，如'jet', 'hot', 'cool', 'magma'等
cmap = plt.get_cmap('viridis')

# 使用imshow来显示矩阵，并通过colormap参数应用颜色映射
plt.imshow(matrix, interpolation='none', cmap=cmap)

# 添加颜色条以表示不同的颜色对应的数值范围
plt.colorbar()

# 显示x轴和y轴的标签
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# 显示标题
plt.title('Matrix Visualization with Color Coding')

# 显示图形
plt.show()
```

```python
import numpy as np

# 初始化一个16x64的矩阵，填充为0
matrix = np.zeros((16, 64))

# 遍历每一行
for pos in range(16):
    # 遍历每一对列（因为sin和cos各占一列）
    for i in range(32):
        # 计算列的索引，注意Python的索引是从0开始的
        col_index = 2 * i
        
        # 计算sin和cos的值，并赋值到矩阵中
        matrix[pos, col_index] = np.sin(pos / 10000 ** (2 * i / 64))
        matrix[pos, col_index + 1] = np.cos(pos / 10000 ** (2 * i / 64))

# 打印矩阵
print(matrix)

# 定义一个颜色映射，这里使用'viridis'，你也可以选择其他颜色映射，如'jet', 'hot', 'cool', 'magma'等
cmap = plt.get_cmap('viridis')

# 使用imshow来显示矩阵，并通过colormap参数应用颜色映射
plt.imshow(matrix, interpolation='none', cmap=cmap)

# 添加颜色条以表示不同的颜色对应的数值范围
plt.colorbar()

# 显示x轴和y轴的标签
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# 显示标题
plt.title('Matrix Visualization with Color Coding')

# 显示图形
plt.show()
```
